```python
def eff(df):
    if (df.annualy_mntn_cost > 1 and df.kilowatt_production < 3):
        return 'high'
    elif (df.annualy_mntn_cost < 1 and df.annualy_mntn_cost > 0.5 and df.kilowatt_production > 3 and df.kilowatt_production < 6):
        return 'high'
    elif (df.kilowatt_production > 6 and df.kilowatt_production < 9):
        return 'high'
    else: 
        return 'Low'
        
```

To manage the quality of vendor, the management wants to control the vendor premium
percentage based on the quality of vendor.
◦ Create a column to adjust the vendor premium(new_vendor_percentage)
▪ 5% increase if
• VendorQuality is “Good”(for instance, 0.3 -> 0.35)
▪ 5% decrease else

```python
def ven_quality(df):
    if (df.ven_quality == 'Good'):
        return df.VendorPercent + 0.05 * df.VendorPercent
    else:
        return df.VendorPercent - 0.05 * df.VendorPercent

df['ven_per']=df.apply(ven_quality,axis=1)
        
```

```python
ven_quality=df.groupby(['region_name','vendor','ven_quality'])[['VendorPercent','ven_per']].mean()
ven_quality
```
